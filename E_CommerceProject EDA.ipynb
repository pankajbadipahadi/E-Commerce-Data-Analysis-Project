```python
import pandas as pd
import plotly.express as px  # It is used for data visualization
import plotly.graph_objects as go  # It is used to create advanced and customize graph
import plotly.io as pio  # It is used for customize graph template
import plotly.colors as colors  # It is used for colors
pio.templates.default ="plotly_white" # It is used for templates
```

```python
df=pd.read_csv('/content/Sample - Superstore.csv',encoding='latin-1')  # Using encoding=latin-1 we read the special character in dataset
df
```

# Converting date **column**

```python
df['Order Date']=pd.to_datetime(df['Order Date'],format='mixed')
df['Ship Date']=pd.to_datetime(df['Ship Date'],format='mixed')
df.info()
```

**Create a month, year & week seperate column**

```python
df['Order Month']=df['Order Date'].dt.month
df['Order Year']= df['Order Date'].dt.year
df['Order Day of Week']= df['Order Date'].dt.dayofweek
df.head()
```

# **Monthly Sales Analysis**

```python
month_sales=df.groupby('Order Month')['Sales'].sum().reset_index()
month_sales
```

```python
fig=px.line(month_sales,
            x='Order Month',
            y='Sales',
            title='Monthly Sales Analysis')
fig.show()
```

```python
df.head()
```

# **Sales by Category**

```python
category_sales=df.groupby('Category')['Sales'].sum().reset_index()
category_sales
```

```python
fig=px.pie(category_sales,
           values='Sales',
           names='Category',
           hole=0.4,
           color_discrete_sequence=px.colors.qualitative.Pastel)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales By Category', title_font=dict(size=24))
fig.show()
```

```python
df.head()
```

# **Sales Analysis by Sub-Category**

```python
sub_category_sales=df.groupby('Sub-Category')['Sales'].sum().reset_index()
sub_category_sales
```

```python
fig=px.bar(sub_category_sales,
           x='Sub-Category',
           y='Sales',
           title='Sales Analysis By Sub-Category')
fig.show()
```

# **Sales and Profit by Customer Segment**

```python
customer_segment=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
customer_segment
```

```python
color_palatte=colors.qualitative.Pastel
fig=go.Figure()
fig.add_trace(go.Bar(x=customer_segment['Segment'],
                     y=customer_segment['Sales'],
                     name='Sales',
                     marker_color=color_palatte[0]))

fig.add_trace(go.Bar(x=customer_segment['Segment'],
                     y=customer_segment['Profit'],
                     name='Profit',
                     marker_color=color_palatte[1]))

fig.update_layout(title='Sales and Profit Analysis by Customer Segment',
                  xaxis_title='Customer Segment',yaxis_title='Amount')

fig.show()
```

```python
df.head()
```

# **Sales to Profit Ratio**

```python
customer_segment=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
customer_segment['Sales_to_Profit_Ratio']= customer_segment['Sales']/customer_segment['Profit']
print(customer_segment[['Segment', 'Sales_to_Profit_Ratio']])
```

```python
fig=px.pie(customer_segment,
           values='Sales_to_Profit_Ratio',
           names='Segment',
           hole=0.4,
           color_discrete_sequence=px.colors.qualitative.Pastel_r)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title_text='Sales To Profit Ratio', title_font=dict(size=24))
fig.show()
```
